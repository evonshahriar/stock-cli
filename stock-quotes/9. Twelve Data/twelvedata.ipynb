<h2>twelvedata: </h2><br><b><i><a href="https://pypi.org/project/twelvedata/" target="_blank">twelvedata</a> is the offical Python client for Twelve Data API.</i></b>

<h2>Installation: </h2><br>
<p>Use the package manager pip to install Twelve Data API library (without optional dependencies):<br>
<code>pip install twelvedata</code><br>
Or install with pandas support:<br>
<code>pip install twelvedata[pandas]</code><br>
Or install with pandas, matplotlib and plotly support used for charting:<br>
<code>pip install twelvedata[pandas,matplotlib,plotly]</code></p>

```python
from twelvedata import TDClient
import sys
import warnings
if not sys.warnoptions:
    warnings.simplefilter("ignore")

td = TDClient(apikey="8393970c7ed745f1bc963a735d340cce")
ts = td.time_series(
    symbol="MSFT",
    outputsize=75,
    interval="1day",
)

# 2. Returns OHLCV + BBANDS(close, 20, 2, SMA) + %B(close, 20, 2 SMA) + STOCH(14, 3, 3, SMA, SMA)
df = ts.with_bbands().with_percent_b().with_stoch(slow_k_period=3).as_pyplot_figure()
```
